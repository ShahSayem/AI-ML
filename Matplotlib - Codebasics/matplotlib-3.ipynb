```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
days = [1, 2, 3, 4, 5, 6, 7]
max_t = [50, 51, 52, 48, 47, 49, 46]
min_t = [43, 42, 40, 44, 33, 35, 37]
avg_t = [45, 48, 48, 46, 40, 42, 41]
```

```python
plt.xlabel("Day")
plt.xlabel("Temp")
plt.title("Weather")

plt.plot(days, max_t, color='red', label='max')
plt.plot(days, min_t, color='green', label='min')
plt.plot(days, avg_t, color='orange', label='avg')

plt.legend(loc='lower left', shadow=True, fontsize='large')
plt.grid()
```
