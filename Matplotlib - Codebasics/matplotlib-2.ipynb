```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [50, 51, 52, 48, 47, 49, 46]
```

```python
plt.plot(x, y, 'g+-.')
```

```python
plt.plot(x, y, 'rD--')
```

```python
plt.plot(x, y, color='red', marker='D', linestyle='')
```

```python
plt.plot(x, y, color='#778800', marker='+', linestyle='', markersize=15)
```

```python
plt.plot(x, y, color='red', linewidth=5, alpha=1)
```

```python
plt.plot(x, y, color='red', linewidth=5, alpha=0.4)
```
