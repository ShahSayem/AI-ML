```python
import pandas as pd
df1 = pd.DataFrame({
    "City": ["Sylhet", "Dhaka", "Chittagong", "Rajshahi"],
    "Temp": [32, 39, 34, 40],
})

df1
```

```python
df2 = pd.DataFrame({
    "City": ["Sylhet", "Rajshahi", "Chittagong"],
    "Humidity": [65, 68, 75]
})

df2
```

```python
df3 = pd.merge(df1, df2, on="City")
df3
```

```python
df4 = pd.merge(df1, df2, on="City", how="outer", indicator=True)
df4
```

```python
df5 = pd.merge(df1, df2, on="City", how="right")
df5
```

```python
df1 = pd.DataFrame({
    "City": ["Sylhet", "Dhaka", "Chittagong", "Rajshahi"],
    "Temp": [32, 39, 34, 40],
    "Humidity": [65, 60, 63, 72]
})

df1
```

```python
df2 = pd.DataFrame({
    "City": ["Sylhet", "Rajshahi", "Chittagong"],
    "Temp": [32, 42, 40],
    "Humidity": [65, 68, 75]
})

df2
```

```python
df_a = pd.merge(df1, df2, on="City")

df_a
```

```python
df_b = pd.merge(df1, df2, on="City", suffixes=("_left", "_right"))

df_b
```
