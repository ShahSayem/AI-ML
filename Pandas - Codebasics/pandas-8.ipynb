```python
import pandas as pd

bd_weather = pd.DataFrame({
    "City": ["Sylhet", "Dhaka", "Chittagong"],
    "Temp": [32, 39, 34],
    "Humidity": [60, 70, 80]
})

bd_weather
```

```python
us_weather = pd.DataFrame({
    "City": ["New York", "Chicago", "Orlando"],
    "Temp": [21, 14, 35],
    "Humidity": [68, 65, 75]
})

us_weather
```

```python
df = pd.concat([bd_weather, us_weather])
df
```

```python
df = pd.concat([bd_weather, us_weather], ignore_index = True)
df
```

```python
df = pd.concat([bd_weather, us_weather], keys = ["bd", "us"])
df
```

```python
df.loc["us"]
```
